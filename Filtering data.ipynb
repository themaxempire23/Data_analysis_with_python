## How to use conditions for Data filtering

#### Filetering 
- selecting certain rows from a dataframe that meet certain conditions, this helps to narrow down the data towards whats relevant for analysis.

```python
import pandas as pd
```

```python
# sanmple dataset
data = {
    'Name': ['Alice', 'Bob', 'Charles', 'David', 'Eva'],
    'Age': [24, 37, 30, 22, 55],
    'Gender': ['F', 'M', 'M', 'M', 'F'],
    'Salary': [5000,60000, 55000, 80000, 45000]
} 

#Defing dataframe
df = pd.DataFrame(data)
print(df)
```

### writing a condition

```python
condition = df['Age'] > 30 # Looking for the people above 30
print(condition)
```

```python
condition = df['Age'] >= 30 # Looking for the people above or equal to 30
print(condition)
```

```python
#Filtering the dataframe
filtered_df = df[condition]
print(filtered_df)
```

### Using multiple conditions

```python
filtered_df = df[(df['Age'] > 25) & (df['Gender'] == 'M')]
print(filtered_df)
```

```python
filtered_df = df[(df['Age'] > 60) & (df['Gender'] == 'M')]
print(filtered_df)
```

## Using query methods (More like sql, suported by pandas)

```python
df.query("Age > 25 and Gender == 'M'")
```
