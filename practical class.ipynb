```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Dataset
data = [7, 3, 14, 9, 7, 8, 13]
```

```python
# box plot visualisation
plt.boxplot(data, vert=False, patch_artist=True)

plt.title("Box-and-Whisker Plot for [7, 3, 14, 9, 7, 8, 13]")
plt.xlabel("Values")
```

## Calculations:

Data sorted in ascending order : [3, 7, 7, 8, 9, 13, 14]

- Minimum = 3
- Q1 = 7
- Median = 8
- Q3 = 13
- Maximum = 14
- This box-plot shows a box from 7 to 13, median at 8, whiskers at 3 and 14.
