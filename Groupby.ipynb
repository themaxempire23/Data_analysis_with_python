## Groupby: 
- grouping or spliting data based on some criteria,  based on categories or values in a coloumn.
  
## Aggregation:
- using functions like mean, sum, max , min etc

```python
import pandas as pd
```

```python
data = {
    'Region':['East', 'West','East','South', 'North', 'North'],
    'Name':['Alice', 'Bob','Charles','David', 'Eva', 'Darwin'],
    'Sales':[200, 150, 300, 100, 400, 600]
}

df = pd.DataFrame(data)
df
```

```python
grouped = df.groupby('Region')
```

```python
# Applying aggregation function
grouped['Sales'].sum()
```

```python
# Applying aggregation function: mean per region
grouped['Sales'].mean()
```

```python
# Applying aggregation function
grouped['Sales'].count()
```

```python
# Using multiple aggretion at once

grouped['Sales'].agg(['sum','mean','max','min'])
```

```python
# Applying aggregation function: Displaying sale per region
df.groupby(['Region', 'Sales'])['Sales'].sum()
```

```python
# resting back to default
grouped['Sales'].sum().reset_index()
```

```python
#sorting the values in the df
grouped = df.groupby('Region')['Sales'].sum()
sorted_group = grouped.sort_values(ascending=False)
print(sorted_group)
```
