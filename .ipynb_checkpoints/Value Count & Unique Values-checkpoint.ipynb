## Value count and unique values in pandas
- used for exploring catergorical data

```python
import pandas as pd
```

```python
data = {
    'City':['New York', 'London', 'New York', 'Paris', 'London', 'Berlin', 'Paris', 'Paris'],
    'Gender':['Male', 'Female','Female','Male', 'Female', 'Male', 'Female', 'Male'],
    #'Sales':[200, 150, 300, 100, 400, 600]
}

df = pd.DataFrame(data)
df
```

```python
# Using value count, determinig how many times a city is applied/ appears
df['City'].value_counts()
```

```python
# Using value count, determinig how many times a Gender is applied/ appears
df['Gender'].value_counts()
```

```python
# Using normalies trues in percentage instead of raw counts
df['Gender'].value_counts(normalize=True)
```

```python
df['City'].value_counts(normalize=True)
```

## showing all unique values in the df

```python
df['City'].unique()
```

```python
df['Gender'].unique()
```

## Determing the exact numbers of the unique values in the dataset

```python
df['Gender'].nunique()
```

```python
df['City'].nunique()
```
