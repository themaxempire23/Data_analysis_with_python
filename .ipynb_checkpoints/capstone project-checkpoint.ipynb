```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('owid-covid-data.csv')
df.head()
```

```python
#finding missing values and filling
df.isnull().sum()
df.fillna(0, inplace=True)
df['Date'] = pd.to_datetime(df['date'])
```

## Eploring the data using EDA

```python
df.describe()
df['location'].value_counts().head(10)
```

```python
top_locations = df[df['date'] == '2020-12-31'].groupby('location')['new_cases'].sum().nlargest(5)
top_locations.plot(kind='bar', title='Top 5 Locations by New cases')
plt.ylabel('Confirmed Cases')
plt.show()
```
