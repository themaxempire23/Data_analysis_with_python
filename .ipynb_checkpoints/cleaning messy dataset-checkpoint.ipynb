```python
import pandas as pd
import numpy as np
```

```python
# sanmple dataset
data = {
    'Name': ['Alice', 'Bob', None, 'David', 'Alice'],
    'Age': ['24', 'thiry', '30', None, '55'], # test scores
    'Salary': [89900, 99090, 609807, 47783, 7878], # test scores
    'Join_date': ['2025-01-01', 'not available', '2025-02-23', '2025-05-07', '2025-01-15']
} 

#Defing dataframe
df = pd.DataFrame(data)
df
```

## viewing Data summary

```python
df.info()
```

```python
df.describe(include='all')
```

## checking for null values in the data

```python
df.isnull().sum()
```

## Dropping or filling the missing values

```python
df['Age'] = df['Age'].replace('thirty', np.nan)
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Age'] = df['Age'].fillna(df['Age'].mean())

df['Name'] = df['Name'].fillna('Unknown')
df['Salary'] = df['Salary'].ffill()

df.describe(include='all')# trying to display what happened
```

```python
df['Join_date'] = pd.to_numeric(df['Join_date'], errors='coerce')
```

## droping duplicates / Getting rid of duplicates

```python
df = df.drop_duplicates()
```

```python
print(df)
print(df.dtypes)
```

## Using the rename function (Optional feature)

```python
df = df.rename(columns={'Names': 'First-Name', 'Salary': 'Salary($)'})
```
