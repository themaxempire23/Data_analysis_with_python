```python
import pandas as pd 
```

```python
# sanmple dataset
data = {
    'Name': ['Alice', 'Bob', 'Charles', 'David'],
    'Age': [24, 37, 30, 22],
    'Score': [85, 90, 78, 100], # Score of a test
    'Country': ['NAM', 'RSA', 'UK', 'BOTS']
} 

#Defing dataframe
df = pd.DataFrame(data)
print(df)
```

## Sorting by single coloumns

```python
# sorting the age in ascending order [small to big]
df_sorted_age = df.sort_values(by='Age')
print(df_sorted_age)
```

```python
# sorting the score in ascending order [small to big]
df_sorted_age = df.sort_values(by='Score')
print(df_sorted_age)
```

## Sorting by multiple coloumns

```python
df_sorted = df.sort_values(by=['Score', 'Age'], ascending=[False, True])
print(df_sorted)
```

## sorting out by index

```python
df_sorted_index = df.sort_index()
print(df_sorted_index)
```

## Indexing
- selecting specific rows and coloumns by selecting labels or position
- Indexing  the way to get an unordered table into an order
- iloc -> position based
- location -> label based

```python
# accessing a row by index location
df.set_index('Name', inplace=True)
print(df.loc['Charles'])
```

```python
# accessing a row by index position
print(df.iloc[3])
```

```python
# accessing specific rows and coloumns
print(df.loc[['Charles', 'Bob', 'Alice'], ['Age', 'Score', 'Country']])
```

```python
# Reseting interger values, going back to default
df_reset = df.reset_index()
print(df_reset)
```
