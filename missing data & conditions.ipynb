```python
import pandas as pd
import numpy as np
```

```python
# sanmple dataset
data = {
    'Name': ['Alice', 'Bob', 'Charles', 'David', 'Eva'],
    'Age': [24, np.nan, 30, 22, np.nan],
    'Gender': ['F', 'M', np.nan, 'M', 'F'],
    'Score': [85,90, np.nan, 66, 76]
} 

#Defing dataframe
df = pd.DataFrame(data)
print(df)
```

Handling missing data

```python
print(df.isnull()) # finding missing data
```

```python
#missing values per coloumn
print(df.isnull().sum())
```

```python
#drop missing values, droping rows with missing values
df_drop_any = df.dropna()
```

```python
df_drop_all = df.dropna(how='all')
```

```python
# droping coloumns with missing values
df_drop_columns = df.dropna(axis=1)
```

Filling missing values

```python
df_fill_constant = df.fillna(0) # fillna replaces values
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['Gender'] = df['Gender'].fillna(df['Gender'].mode())
```

```python
# Forward filling -> filling with previous values
print(df)
```

```python
df_ffill = df.ffill() # -> forward fill
```

```python
df_ffill = df.bfill() # -> backward fill
```

```python
## interpolating missing data / changing string values to numerical

df[df.select_dtypes(include=['number']).columns] = df.select_dtypes(include=['number']).interpolate
```

```python
# Checking if all missing values are handled
print(df.isnull().sum())
```
