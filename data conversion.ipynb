## Data type Conversion
- Converting data from one format to another format

```python
import pandas as pd
```

```python
# sanmple dataset
data = {
    #'Name': ['Alice', 'Bob', 'Charles', 'David', 'Eva'],
    'Age': ['24', '37', '30', '22', '55'], # test scores
    'Salary': [899000, 99090, 6098807, 478783, 787888], # test scores
    'Join_date': ['2025-01-01', '2025-05-03', '2025-02-23', '2025-05-07', '2025-01-15']
} 

#Defing dataframe
df = pd.DataFrame(data)
print(df)
```

```python
# Checking original datatypes
df = pd.DataFrame(data)
print(df.dtypes)
```

## Converting the Age strings into intergers

```python
df['Age'] = df['Age'].astype(int)
```

```python
# converting the dates
df['Join_date'] = pd.to_datetime(df['Join_date'])
```

```python
print(df.dtypes)
```

## Advanced convertion features and techniques

```python
#adding new coloumns for practice
df['Height'] = ['5.3', '6.7', '8.9', '5.4', '65.3']
df['Department'] = ['IT', 'HR', 'Finance', 'Procurement', 'Exco']
df['Experience'] = ['4', 'three', 'nine', 'twenty', '50']

print(df)
```

```python
df['Height'] = df['Height'].astype(float)
```

```python
print(df.dtypes)
```

```python
## Converting department into catergory
df['Department'] = df['Department'].astype('category')
```

```python
print(df.dtypes)
```

```python
## Converting experience into numerics
df['Experience'] = pd.to_numeric(df['Experience'], errors='coerce')
```

```python
print(df.dtypes)
```

## end product deisplaying everything, after conversion

```python
print(df)
print("\nData types:\n", df.dtypes)
```
