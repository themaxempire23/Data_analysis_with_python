```python
import pandas as pd
import numpy as np
```

```python
# sanmple dataset
data = {
    'Name': ['Alice', 'Bob', 'Charles', 'David', 'Eva'],
    'Math': [24, 37, 30, 22, 55], # test scores
    'Science': [89, 90, 67, 43, 78] # test scores
} 

#Defing dataframe
df = pd.DataFrame(data)
print(df)
```

## creating new coloumns based on calculated values as average

```python
df['Average'] = (df['Math'] + df['Science']) / 2 # getting calculated average mark
print(df)
```

## creating new coloumn based on condition, either pass or fail

```python
df['Results'] = df['Average'].apply(lambda x: 'Pass' if x > 50 else 'Fail')
print(df)
```

## creating a column with constant value

```python
df['Year'] = 2025
print(df)
```

## creating a cloumn to assing the grade

```python
def grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    else:
        return 'C'
    if score <= 50:
        return 'D'    


df['Grade'] = df['Average'].apply(grade)
print(df)
```

## Creating a coloumn called math levele

```python
df['Math_Level'] = df['Math'].apply(lambda x: 'High'if x > 80 else 'Low')
print(df)
```

## Creating new coloumn -> Science_Level
- How to use np.where

```python

df['Science_Level'] = np.where(df['Science'] >= 90, 'Top', 'Average')
print(df)
```

## Creating new coloumn called grade description
- How to use the map function

```python
df['Grade'] =  df['Math'].apply(lambda x: 'A' if x >= 90 else ('B' if x >= 80 else 'C'))

grade_map = {
    'A': "Execelent",
    'B': "Good",
    'C': "Can do better",
    'D': "Needs improvement",
}

df['Grade_Desc'] = df['Grade'].map(grade_map)
print(df)
```
